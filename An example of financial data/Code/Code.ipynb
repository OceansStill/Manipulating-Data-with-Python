<strong>3.  </strong>
- 统计**各指标**的以下信息
    - 起始时间（第一个非空时间点，列名命名为`BegDt`）
    - 终止时间（最后一个非空时间点，列名命名为`EndDt`）
    - 观测总数（第一个非空时间点 至 最后一个非空时间点中 的月份个数，列名命名为`Obs`）
    - 缺失值个数和相应的个数占比（第一个非空时间点至最后一个非空时间点中缺失值的个数，列名分别命名为`NaNNum`、`NaNPct`）
    - 负值个数、相应的个数占比、和负值对应的时间点（第一个非空时间点至最后一个非空时间点中负数的个数，列名分别命名为`NegNum`、`NegPct`、`NegLst`）
- 并将统计好的结果赋值给变量`df_summary_1`。可参考变量`df_summary_1` 的内容为如下形式（前文所引括号为各列的列名）

```python
BegDt = df.apply(lambda series: series.first_valid_index())
EndDt=df.apply(lambda series: series.last_valid_index())
Obs=(EndDt).dt.to_period('M').astype(int)-BegDt.dt.to_period('M').astype(int)+1
NaNNum=df.apply(lambda series:series[series.first_valid_index():series.last_valid_index()].isna().sum() )
NaNPct=round(NaNNum/Obs*100,2)
NegNum=df.apply(lambda series:(series[series.first_valid_index():series.last_valid_index()]<0).sum() )
NegPct=round(NegNum/Obs*100,2)
index=df.index
df.index=pd.Series(df.index.values).dt.to_period('M')
NegLst=df.apply(lambda series:series[series.first_valid_index():series.last_valid_index()].index
[series[series.first_valid_index():series.last_valid_index()]<0 ].values )
df_summary_1=pd.DataFrame({"BegDt":BegDt,"EndDt":EndDt,"Obs":Obs,"NaNNum":NaNNum,"NaNPct":NaNPct,
"NegNum":NegNum,"NegPct":NegPct,"NegLst":NegLst})
df.index=index
#df_summary_1["NegPct"]=df_summary_1["NegPct"].apply(lambda list: )
df_summary_1
```

<strong>1.  </strong>请使用你认为的最简洁的方式：对于数据框df1中的各指标，使用各指标上一期的值对当期为缺失的进行填充，并将填补后的数据框赋值给变量df2

```python
df2=df1.fillna(df1.median())
df2
```

<strong>2.  </strong>
请使用**你认为的最简洁**的方式：由于各列指标中包含一些负值，故需要对数据框`df2`中的各指标逐列进行正向化操作（**Min-Max normalization**），使用的公式为：

$$
x' = \dfrac{x - \min \{ x  \}}{\max \{x\} - \min \{ x \}} + 1
$$

    并将正向化后的数据框赋值给变量`df3`

```python
df3=(df2-df2.min())/(df2.max()-df2.min())+1
df3
```

<strong>3.  </strong>请使用**你认为的最简洁**的方式：使用数据框`df3`，先计算得到各指标每月的**同比变动**，然后使用`pandas`**处理缺失值的相关函数**剔除包含`NaN`的月份，并将处理好的数据框赋值给变量`df_pct`
    - Note
        - 指标同比变动的计算公式为：$\dfrac{某指标今年的当期值}{该指标上一年度的当期值} - 1$

```python
df_pct=(df3/df3.shift(12)-1).dropna(axis=0)
df_pct.index=pd.Series(df_pct.index.values).dt.to_period('M')

df_pct
```
